```python
import random
from openai import OpenAI

client = OpenAI()

def roll_dice():
    return random.randint(1, 6)
```

```python
target = 6
max_iterations = 20
```

```python
history = ""
iteration = 0

while iteration < max_iterations:
    iteration += 1
    
    instructions = f"""
    You are a dice-rolling agent. Your goal is to roll a {target} using a standard six-sided die. You can roll the die as many times as needed, but you have a maximum of {max_iterations} rolls. After each roll, decide your next action Based on what you have rolled so far.
    
    You've made the following rolls so far: {history if history else "No rolls yet."}
    
    Option A: If you have rolled the {target}, stop rolling and respond with "Success!".
    Option B: If you have not rolled the {target}, try rolling the {target} again and respond "Rollin' again!".
    """
    
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[
            {"role": "system", "content": instructions}
        ]
    )
    
    result = response.choices[0].message.content
    
    if "Success!" in result:
        print(result)
        break
        
    if "Rollin' again!" in result:
        roll = roll_dice()
        print(f"Roll {iteration}: {roll}")
        history += f"Roll {iteration}: {roll}\n"
        
print(f"Game ended after {iteration-1} interations.")
        
```
