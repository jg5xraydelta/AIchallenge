### Step 1: Import API keys and libraries

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown, display


load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")

if OPENAI_API_KEY is None:
    raise Exception("API_KEY is missing.")
else:
    print(OPENAI_API_KEY[:8])
    
client = OpenAI()
```

### Step 2: UI with AI

```python
import gradio as gr

def respond_ai(message, history):
    messages = [{"role": "system", "content": "You are a helpful assistant"}] + history + [{"role": "user", "content": message}]
    client = OpenAI(api_key=OPENAI_API_KEY)
    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=messages
    )
    
    reply = response.choices[0].message.content
    return reply

gr.ChatInterface(fn=respond_ai).launch(inbrowser=True, inline=False)
    
```
