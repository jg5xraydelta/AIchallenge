```python
### Step 1: Import libraries and API Keys
```

```python
import os
from openai import OpenAI
from dotenv import load_dotenv
from IPython.display import Markdown

load_dotenv()

OPENAI_API_KEY = os.getenv("OPENAI_API_KEY")
client = OpenAI()
```

### Step 2: Searching the web.

```python
client = OpenAI()

completion = client.chat.completions.create(
    model="gpt-4o-mini-search-preview",
    messages=[
        {"role": "system", "content": "What is today's date? Tell me the top stories in AI in this past week according to reliable media.  Each story should be from a separate source and date. Include dates."}
    ],
    max_tokens=500
) 
    
```

```python
search_results = completion.choices[0].message.content
display(Markdown(search_results))
```

### Step 3: Summarize and write social media post.

```python
system_message = """
You are a helpful assistant that creates amazing lindein posts about AI based on 
text that you are provided.  Make the posts engaging and interesting.  
ALso make it concise and a thought provoking question at the end to encourage comments.
"""

prompt = f"""
The following text contains top 3 news stories about AI in the past week.
Now please write my linkedin post based on this information.property

{search_results}
"""
completion = client.chat.completions.create(
    model="gpt-4o-mini",
    messages=[
        {"role": "system", "content": system_message},
        {"role": "user", "content": prompt}
    ],
    max_tokens=500,
    temperature=0.7
)

chat_response = completion.choices[0].message.content
```
